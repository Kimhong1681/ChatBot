```python
import asyncio
import nest_asyncio
from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import Application, CommandHandler, CallbackQueryHandler

# Apply nest_asyncio to prevent event loop errors in Jupyter Notebook
nest_asyncio.apply()

# Bot token
TOKEN = " 8116272316:AAHI5jabceaEiqCsIMAmFwp-Kv-Fj6g21kE"

# Define categories, questions, and answers for English and Khmer
CATEGORIES = {
    "kh": {  # Khmer categories
        "category1": "អំពីស្ថាប័ណ្ណ",
        "category2": "លក្ខខណ្ឌរបស់អតិថិជន",
        "category3": "ពាក្យសុំឥណទាន",
        "category4": "ការសង់ប្រាក់កម្ចី"
    },
    "en": {  # English categories
        "category1": "About Us",
        "category2": "Term and Condition",
        "category3": "How to Request",
        "category4": "Pay Back"
    }
}

QUESTIONS = {
    "kh": {  # Khmer questions
        "category1": {
            1: "តើស្ថាប័ណ្ណខាងបងជាក្រុមហ៊ុន ឬក៏ជាមីក្រូ ឬក៏ជាធនាគារ?",
            2: "តើកម្ចី 8Stars Finance Digital Loan គឺជាអ្វី?",
            3: "ហេតុអ្វីបានជាខ្ញុំត្រូវជ្រើសរើសប្រើប្រាស់កម្ចី 8Stars Finance ឌីជីថលឡូន?", 
            4: "តើខ្ញុំអាចដឹងថាសំណើកម្ចីរបស់ខ្ញុំត្រូវបានអនុម័តដោយរបៀបណា?",
            5: "បើខ្ញុំមានសំណួរផ្សេងទៀតអំពីប្រាក់កម្ចីឌីជីថល 8Stars Finance តើខ្ញុំគួរទំនាក់ទំនងទៅអ្នកណា?"
        },
        "category2": {
            6: "តើតម្រូវឲ្យមានឯកសារអ្វីខ្លះដើម្បីស្នើសុំកម្ចី?",
            7: "តើ អ្នកណាដែលមានសិទ្ធដាក់ពាក្យសុំប្រាក់កម្ចីឌីជីថល 8Stars Finance?"
        },  
        "category3": {
            8: "ខ្ញុំមិនចេះស្នើកម្ចីដោយខ្លួនឯងទេ តើអ្នកណាអាចជួយស្នើឲ្យខ្ញុំបានដែរឬទេ?",
            9: "តើខ្ញុំអាចស្នើកម្ចី 8Stars Finance ឌីជីថលឡូនបានដោយរបៀបណា?",
            10: "តើខ្ញុំអាចស្នើកម្ចីជាលើកទី 2 បានដែរឬទេ?"
        },
        "category4": {
            11: "តើខ្ញុំអាចបង់សងប្រចាំខែតាមរបៀបណាខ្លះ?",
            12: "តើខ្ញុំអាចបង់ផ្ដាច់កម្ចីនេះតាមរបៀបណា?"

        }
    },
    "en": {  # English questions
        "category1": {
            1: "Is your institution a company, a microfinance institution, or a bank?",
            2: "What is 8Stars Finance Digital Loan?",
            3: "Why should I choose to apply for an 8Stars Finance Digital Loan?",
            4: "How will I know if I'm Approved? How can I check my loan application status?",
            5: "If I have more questions about 8Stars Finance Digital Loan, who should I contact?",

        },
        "category2": {
            6: "What documents are required to apply for the loan?",
            7: "Who is eligible to apply for the 8Stars Finance Digital Loan?"
        },
        "category3": {
            8: "I don't know how to apply, can I ask someone to apply for me?",
            9: "How can I apply for an 8Stars Finance Digital Loan?",
            10: "Is it possible to have more than one loan at a time?"
        },
        "category4": {
            11: "1.	How can I pay for the monthly installment?",
            12: "How can I pay off the loan?"
    }
}
}

ANSWERS = {
    "kh": {  # Khmer answers
        1: "យើងជាគ្រឹះស្ថានមានអាជ្ញារប័ណ្ណក្រោមធនាគារជាតិនៃកម្ពុជា។",
        2: "កម្ចី 8Stars Finance Digital Loan គឺជាកម្ចីឌីជីថិល",
        3: "ព្រោះកម្ចី 8Stars Finance ឌីជីថិលឡូនមានភាពងាយស្រួល រហ័ស និងអាចទទួលបានការអនុមតភ្លាមៗ ដោយមិនចាំបាច់រង់ចាំយូរ។ ",
        4: "អ្នកនឹងទទួលបានសារជូនដំណឹងតាមរយៈកាហៅទូរស័ព្ទពីបុគ្គលិក ឬសារ (SMS) នៅពេលដែលកម្ចីត្រូវបានអនុមត។",
        5: "អ្នកអាចទំនាក់ទំនងទៅសេវាកម្មអតិថិជន 8Stars Finance តាមរយៈ៖ 023 934444 និងប្រព័ន្ធ ផ្សព្វផ្សាយ សង្គមរបស់យើង។",
        6: "តម្រូវអោយមាន ឯកសារចម្បងមួយចំនួន ដូចជារូបថតអត្តសញ្ញាណប័ណ្ណនិងលិខិតបញ្ជាក់ប្រាក់ចំណូលរបស់អ្នក។",
        7: "អតិថិជនទាំងអស់ដែលមានអាយុចាប់ពី 18 - 65 ឆ្នាំ។",
        8: "ប្រាកដណាស់ អ្នកអាចទាក់ទងទៅលេខ 023934444 ឬមើលតាមរយះ វីឌីអូដើម្បីស្នើកម្ចីបាន។",
        9: "អ្នកអាចស្នើកម្ចី 8Stars Finance ឌីជីថលឡូននេះបាន តាម 8Stars Finance Bots/ App, ។ សម្រាប់ព័ត៌មានលម្អិតពីរបៀបស្នើសុំ សូមចុច Link..",
        10: "អ្នកអាចមានកម្ចីតែមួយក្នុងពេលតែមួយ។",
        11: "អ្នកអាចបង់សងត្រឡប់តាមរយះ KHQR   or Wing, TrueMoney ",
        12: "អ្នកអាចបង់ផ្ដាច់កម្ចីឌីជីថលឡូននេះបានតាមអេប KHQR& 8Stars Finance, Wing, True Money."
    },
    "en": {  # English answers
        1: "We are a Microfinance institution licensed under the National Bank of Cambodia.",
        2: "8Stars Finance Digital Loan is an online loan service.",
        3: "Because 8Stars Finance Digital Loan are easy, fast, and can be approval and disbursement instantly. No waiting time.",
        4: "You will receive a phone call from staff, notification, or SMS when your loan is approved. You can also check your application or loan status by tapping 8Stars Finance Digital Loan function in 8Stars Finance Bots/ App.",
        5: "You can contact 8Stars Finance Customer Service via 023 934444 and our social media platforms.",
        6: "តម្រូវអោយមាន ឯកសារចម្បងមួយចំនួន ដូចជារូបថតអត្តសញ្ញាណប័ណ្ណនិងលិខិតបញ្ជាក់ប្រាក់ចំណូលរបស់អ្នក។",
        7: "All customers between the ages of 18 and 65 are eligible",
        8: "Yes, you can call 023934444 or Check video via link to apply for the loan",
        9: "You can apply Digital Loan via 8Stars Finance Bots/App, for more details on how to apply loan please click here",
        10: "You are only eligible for one loan at a time.",
        11: "You can pay for the monthly installment via KHQR   or Wing, TrueMoney",
        12: "You can pay off the loan via KHQR& 8Stars Finance, Wing, TrueMoney"
    }
}

# ----------------------------------------------
# Functions to handle the interaction for Khmer and English
# ----------------------------------------------

async def start(update: Update, context):
    """Start command: Ask user to select language."""
    keyboard = [
        [InlineKeyboardButton("ភាសាខ្មែរ", callback_data="lang_kh")],
        [InlineKeyboardButton("English", callback_data="lang_en")]
    ]
    await update.message.reply_text(
        "សួស្តី! សូមជ្រើសរើសភាសារបស់អ្នក ❤️:" if context.user_data.get("language", "en") == "kh" else "Hello! Please choose your language ❤️:",
        reply_markup=InlineKeyboardMarkup(keyboard)
    )

async def language_selection(update: Update, context):
    """Handle language selection and show categories."""
    query = update.callback_query
    lang = query.data.split("_")[1]  # Extract 'kh' or 'en'

    context.user_data["language"] = lang
    context.user_data["history"] = []  # Reset history
    context.user_data["category"] = None  # Reset category

    await display_categories(query, context, lang)

async def display_categories(query, context, lang):
    """Display categories to the user."""
    categories = CATEGORIES[lang]
    keyboard = [
        [InlineKeyboardButton(cat, callback_data=f"cat_{cat_key}")] 
        for cat_key, cat in categories.items()
    ]

    await query.edit_message_text(
        "សូមជ្រើសរើសប្រភេទ:" if lang == "kh" else "Choose a category:",
        reply_markup=InlineKeyboardMarkup(keyboard)
    )

async def category_selection(update: Update, context):
    """Handle category selection and display questions."""
    query = update.callback_query
    lang = context.user_data.get("language", "en")
    category_key = query.data.split("_")[1]

    context.user_data["category"] = category_key
    await display_questions(query, context, lang, category_key)

async def display_questions(query, context, lang, category_key):
    """Display questions for the selected category, excluding already answered questions."""
    questions = QUESTIONS[lang].get(category_key, {})
    
    # Get already answered question IDs from the user's history
    answered_questions = set(context.user_data.get("history", []))
    
    # Filter out questions that have already been answered
    remaining_questions = {q_id: q for q_id, q in questions.items() if q_id not in answered_questions}
    
    # Prepare the keyboard
    keyboard = [
        [InlineKeyboardButton(q, callback_data=f"q_{q_id}")] 
        for q_id, q in remaining_questions.items()
    ]

    # If there are no remaining questions, offer the end option
    if not remaining_questions:
        keyboard.append([InlineKeyboardButton("End", callback_data="end")])

    await query.edit_message_text(
        "Choose a question:" if lang == "en" else "សូមជ្រើសរើសសំណួរ:",
        reply_markup=InlineKeyboardMarkup(keyboard)
    )

async def button_handler(update: Update, context):
    """Handle question selection and show answers."""
    query = update.callback_query
    lang = context.user_data.get("language", "en")
    question_id = int(query.data.split("_")[1])

    # Track the question ID in history
    context.user_data.setdefault("history", []).append(question_id)

    # Prepare the response text
    response_text = ""
    for q_id in context.user_data["history"]:
        response_text += f"**Q:** {QUESTIONS[lang].get(context.user_data['category'], {}).get(q_id, '')}\n**A:** {ANSWERS[lang].get(q_id, '')}\n\n"

    # Provide options to go back or end
    keyboard = [
        [InlineKeyboardButton("<< Go Back", callback_data="back")],
        [InlineKeyboardButton("End", callback_data="end")]
    ]

    await query.edit_message_text(response_text.strip(), reply_markup=InlineKeyboardMarkup(keyboard))
    await query.answer()

async def back_to_menu(update: Update, context):
    """Return to the question menu for the current category."""
    query = update.callback_query
    lang = context.user_data.get("language", "en")
    
    # Get the current category from user data
    category_key = context.user_data.get("category", None)
    if category_key:
        await display_questions(query, context, lang, category_key)
    else:
        # If no category is selected, display categories again (as fallback)
        await display_categories(query, context, lang)

async def end_handler(update: Update, context):
    """End the conversation."""
    query = update.callback_query
    lang = context.user_data.get("language", "en")

    await query.edit_message_text(
        "Thank you! The conversation has ended. ❤️" if lang == "en" else "សូមអរគុណ! ការសន្ទនាបានបញ្ចប់។ ❤️",
        reply_markup=InlineKeyboardMarkup([[InlineKeyboardButton("Start Over", callback_data="start_over")]])
    )

async def start_over(update: Update, context):
    """Handle the 'Start Over' action and go back to the language selection."""
    # Reset user data, including question history, selected category
    context.user_data.clear()  # Clear all user data to reset the conversation

    # Send the language selection message again, like a new start
    keyboard = [
        [InlineKeyboardButton("ភាសាខ្មែរ", callback_data="lang_kh")],
        [InlineKeyboardButton("English", callback_data="lang_en")]
    ]
    
    lang = "kh"  # Default language for the new start (can be modified)
    await update.callback_query.edit_message_text(
        "សួស្តី! សូមជ្រើសរើសភាសារបស់អ្នក ❤️:" if lang == "kh" else "Hello! Please choose your language ❤️:",
        reply_markup=InlineKeyboardMarkup(keyboard)
    )

# Main Application
async def main():
    """Set up and run the bot."""
    app = Application.builder().token(TOKEN).build()

    app.add_handler(CommandHandler("start", start))  # Start command
    app.add_handler(CallbackQueryHandler(language_selection, pattern="^lang_"))  # Language selection
    app.add_handler(CallbackQueryHandler(category_selection, pattern="^cat_"))  # Category selection
    app.add_handler(CallbackQueryHandler(button_handler, pattern="^q_"))  # Question selection
    app.add_handler(CallbackQueryHandler(back_to_menu, pattern="^back$"))
    app.add_handler(CallbackQueryHandler(end_handler, pattern="^end$"))
    app.add_handler(CallbackQueryHandler(start_over, pattern="^start_over$"))  # Start over handler

    print("Bot is running...")
    await app.run_polling()

if __name__ == "__main__":
    try:
        asyncio.run(main())
    except RuntimeError:
        loop = asyncio.get_event_loop()
        loop.run_until_complete(main())
```
